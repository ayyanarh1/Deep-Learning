```python
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_data = ImageDataGenerator(rescale=1/255,shear_range=0.2,zoom_range=0.2, horizontal_flip=True,vertical_flip=True )   #rescaling images to 1 and 0
```

```python
test_data = ImageDataGenerator(rescale=1/255)
```

```python
train_data_generator = train_data.flow_from_directory(directory = "/content/drive/MyDrive/Colab Notebooks/DL_Day 7 CNN/Train_Data",target_size=(256,256))
```

```python
test_data_generator = train_data.flow_from_directory(directory = "/content/drive/MyDrive/Colab Notebooks/DL_Day 7 CNN/Test_Data",target_size=(256,256))
```

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
```

```python
model=Sequential()
model.add(Conv2D(input_shape=(256,256,3),filters=8 ,kernel_size=(3,3),activation="relu",padding="same",strides=(1,1)))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=8,kernel_size=(3,3),activation="relu",padding="same",strides=(1,1)))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=16,kernel_size=(3,3),activation="relu",padding="same",strides=(1,1)))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
model.add(Flatten())
model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=3,activation="softmax"))
model.summary()
```

```python
model.compile(optimizer="rmsprop",loss="categorical_crossentropy",metrics=["accuracy"])
```

Lets train our model

```python
model=model.fit(x=train_data_generator,validation_data=test_data_generator,epochs=10)
```
