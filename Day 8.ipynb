```python
from tensorflow import keras
from keras.datasets import mnist  # this is to load the data
from keras.models import Sequential
from keras.layers import Conv2D, AveragePooling2D,Flatten,Dense
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train.shape  # this is training data 60,000 for training, note Lenet operation requires 32x32, so we need to padding operation
```

Before applying padding, we need to ensure `x_train` contains the correct image data, as it was previously corrupted. We will reload the data and then apply the padding.

```python
# Reload the MNIST data to ensure x_train is in its original state
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Apply padding to x_train
x_train = np.pad(array=x_train, pad_width=((0, 0), (2, 2), (2, 2))) # pad width i give zeros(0,0) ,both starting ending column(1,1) and row(1,1)

# Display the new shape to confirm padding
print(x_train.shape)
```

```python
x_test = np.pad(array=x_test, pad_width=((0, 0), (2, 2), (2, 2)))
print(x_test.shape)
```

```python
import pandas as pd
pd.DataFrame(x_train[3000])
```

```python
y_train[3000]
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
pd.DataFrame(x_train[3000]) # it is converting everything between zero to 1
```

```python
from keras.models import Sequential
from keras.layers import Conv2D

model = Sequential()
model.add(Conv2D(input_shape=(32,32,1),filters=6,kernel_size=(5,5),activation='tanh'))
model.add(AveragePooling2D(pool_size=(2,2),strides=2)) # if give strides= none it takes default values

model.add(Conv2D(filters=16,kernel_size=(5,5),activation='tanh'))
model.add(AveragePooling2D(pool_size=(2,2),strides=2))

model.add(Conv2D(filters=120,kernel_size=(5,5),activation='tanh'))

model.add(Flatten())

model.add(Dense(units=84,activation='tanh'))
model.add(Dense(units=10,activation='softmax'))


model.summary()
```

Compile our model

```python
model.compile(optimizer='rmsprop',loss='sparse_categorical_crossentropy',metrics=['sparse_categorical_accuracy'])
```

```python
model.train = model.fit(x=x_train,y=y_train,epochs=15,validation_data=(x_test,y_test))
```

```python
model.evaluate(x_test,y_test)
```

```python
predict=model.predict(x_test)
```
