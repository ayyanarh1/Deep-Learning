```python
x_train=x_train/255 #255 is pixel range
x_test=x_test/255
```

# 1. Build a marchitecture
# 2. Model compilation
# 3. Train our model
#4. Model testing & eveluation

```python
from keras.models import Sequential    # sequencial helps to staock the layes on on another
from keras.layers import Dense,Flatten # Dense layaer is responsible to build the layer, also before giving it to nural network i need to flatten the image, so i import that as well
```

Building model, first create instance of the model

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28))) # i use add funtion and flatern the image before giving it to nural network
# after flattering we need to add hidden layer as below by using dense
model.add(Dense(units=50,activation='relu')) # in the input layer we have 784 nurons, now next layer(hidden) we have 50(randomly given)now. also give acti6vation funtion
model.add(Dense(units=10,activation='relu'))
# now i go with out put layer
model.add(Dense(units=10,activation='softmax')) # units are nurons
model.summary()  # the 784 is nurons
```

Now we need to compile our model, take optimizer/loss and metriccs from keras web

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['sparse_categorical_accuracy'])
```

```python
model_training = model.fit(x=x_train,y=y_train,batch_size=10,epochs=20, verbose='auto') # batch size is number of images, by default it is 32, the total number of images is 60000
# the current iteration we gave 10, suppose for 60000 it may take 60000 iterations, epochs is nothing but all the images tested 20 times
```

```python
model_training = model.fit(x=x_train,y=y_train,batch_size=32,epochs=5, verbose=3,validation_data=(x_test,y_test)) # verbos is howmuch information you want visualise
```

Now we predict the model

```python
y_pred = model.predict(x_test)
```
