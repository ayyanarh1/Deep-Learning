```python
def create_model(neuron1,neuron2,activation_function,init,learning_rate,dropout_rate):
    model = Sequential()

    model.add(Dense(unite=neuron1,activation=activation_function,kernel_initializer=init,input_dim=8))
    model.add(Dropout(dropout_rate))
    model.add(Dense(unite=neuron2,activation=activation_function,kernel_initializer=init))
    model.add(Dropout(dropout_rate))
    model.add(Dense(unite=1,activation='sigmoid',kernel_initializer='glorot_uniform'))
    model.summary()

    adam_opt=tf.keras.optimizers.Adam(learning_rate=learning_rate,beta_1=0.5,beta_2=0.96,epsilon=1e-07)
    model.compile(optimizer=adam_opt,loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=create_model)
```
